# Derivative intuition


Definition of derivative:



This is the derivative of some function $f$ at a point $a$, or $f'(a)$, or $\dfrac{df}{dx}(a)$.

```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
  return 3*x**2 - 4*x + 5
```

```python
xs = np.arange(-5, 6, 0.25)
ys = f(xs)
plt.plot(xs, ys)
```

```python
a = 3
h = .0001
(f(a+h) - f(a)) / h
```

```python
def f(x, y, z):
    return x * y + z
a = 2.0
b = -3.0
c = 10.0
d1 = f(a, b, c)
d2 = f(a+h, b, c)
print(d1)
print(d2)
print((d2-d1)/h)
```
